```python
import numpy as np
import pandas as pd
import random
from sklearn.utils import shuffle
```

```python
n = 1000
```

```python
ketinggian_kentang = [random.randint(300,3000) for i in range(n)]
hujan_kentang = [random.randint(2400,3600) for i in range(n)]
temperatur_kentang = [random.randint(13,24) for i in range(n)]
```

```python
ketinggian_bayam = [random.randint(5,2000) for i in range(n)]
hujan_bayam = [random.randint(1000,2000) for i in range(n)]
temperatur_bayam = [random.randint(20,30) for i in range(n)]
```

```python
ketinggian_stroberi = [random.randint(1000,1500) for i in range(n)]
hujan_stroberi = [random.randint(600,700) for i in range(n)]
temperatur_stroberi = [random.randint(17,20) for i in range(n)]
```

```python
ketinggian_jambuair = [random.randint(5,500) for i in range(n)]
hujan_jambuair = [random.randint(500,3000) for i in range(n)]
temperatur_jambuair= [random.randint(10,28) for i in range(n)]
```

```python
ketinggian_cabai = [random.randint(5,1400) for i in range(n)]
hujan_cabai = [random.randint(800,2000) for i in range(n)]
temperatur_cabai= [random.randint(18,30) for i in range(n)]
```

```python
ketinggian = []
hujan = []
temperatur = []
```

```python
ketinggian.extend(ketinggian_kentang)
ketinggian.extend(ketinggian_bayam)
ketinggian.extend(ketinggian_stroberi)
ketinggian.extend(ketinggian_jambuair)
ketinggian.extend(ketinggian_cabai)
```

```python
hujan.extend(hujan_kentang)
hujan.extend(hujan_bayam)
hujan.extend(hujan_stroberi)
hujan.extend(hujan_jambuair)
hujan.extend(hujan_cabai)
```

```python
temperatur.extend(temperatur_kentang)
temperatur.extend(temperatur_bayam)
temperatur.extend(temperatur_stroberi)
temperatur.extend(temperatur_jambuair)
temperatur.extend(temperatur_cabai)
```

```python
tanaman = []
tanaman.extend(["kentang" for i in range(n)])
tanaman.extend(["bayam" for i in range(n)])
tanaman.extend(["stroberi" for i in range(n)])
tanaman.extend(["jambu air" for i in range(n)])
tanaman.extend(["cabai" for i in range(n)])
```

```python
df = pd.DataFrame(list(zip(tanaman, ketinggian, hujan, temperatur)), columns=['tanaman', 'ketinggian', 'hujan', 'temperatur'])
```

```python
df
```

```python
df2 = shuffle(df)
```
